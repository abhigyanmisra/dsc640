```python
# Import required libraries
import csv
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
# Read world population data
fileData1 = 'data/world-population.xlsm'
population = pd.read_excel(fileData1, sheet_no = 1)

# Display head
population.head()
```

### World Population data
#### Line and Step Charts

```python
# Assigning Values to plot from the dataframe
X = population['Year']
Y = population['Population']


# Line Plot
plt.figure(figsize=(15,10))
plt.title("Population Change with Year")
plt.xlabel("Year")
plt.ylabel("Population in Billions")
plt.grid() ## Adding Grid
plt.plot(X, Y/1000000000) # Dividing population by billion to display values in billions
plt.show()
```

```python
# Step plot
plt.figure(figsize=(15,10))
plt.title("Population Change with Year")
plt.xlabel("Year")
plt.ylabel("Population in Billions")
plt.grid() ## Adding Grid
plt.step(X, Y/1000000000) # Dividing population by billion to display values in billions
plt.show()
```
