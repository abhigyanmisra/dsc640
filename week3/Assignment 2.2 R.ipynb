```r
# Loading required packages
library(magrittr)
library(readxl)
library(ggplot2)
library(dplyr)
library(pastecs)

# Import data to be used for visualization
population = read_excel(path = 'data/world-population.xlsm')

# Examine data
head(population)
```

### World Population data
#### Line and Step graph

```r
# Setting the chart size
options(repr.plot.width = 15, repr.plot.height = 10)

# plotting the line graph for population change with year
# divided population by billion, to reflect values in billions on y-axis
ggplot2::ggplot(data=population, ggplot2::aes(x=Year, y=Population/1000000000)) +
  ggplot2::geom_line(linetype='solid', color='blue', size=1.2) +
  ggplot2::ylab('Population in Billions') +
  ggplot2::ggtitle('Population Change with Year') +
  ggplot2::theme(plot.title = element_text(size = 35),
        axis.title.x = element_text(size = 25),
        axis.title.y = element_text(size = 25),
        axis.text.x = element_text(color = "grey20", size = 20, angle = 0, hjust = .5, vjust = .5, face = "plain"),
        axis.text.y = element_text(color = "grey20", size = 20, angle = 0, hjust = .5, vjust = .5, face = "plain"))
```

```r
# Step plot
# divided population by billion, to reflect values in billions on y-axis
ggplot2::ggplot(data=population, ggplot2::aes(x=Year, y=Population/1000000000)) +
  ggplot2::geom_step(linetype='solid', color='blue', size=1.2) +
  ggplot2::ylab('Population in Billions') +
  ggplot2::ggtitle('Population Change with Year') +
  ggplot2::theme(plot.title = element_text(size = 35),
        axis.title.x = element_text(size = 25),
        axis.title.y = element_text(size = 25),
        axis.text.x = element_text(color = "grey20", size = 20, angle = 0, hjust = .5, vjust = .5, face = "plain"),
        axis.text.y = element_text(color = "grey20", size = 20, angle = 0, hjust = .5, vjust = .5, face = "plain"))
```
